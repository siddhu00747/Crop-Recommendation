#### crop ko chose krke oski min max mean nikalna

```python
@interact  ## isse ek drakdown sa bn jata hai jisse hm select kr skte hai baar baar fun ko call nhi krna pdta
def summary(crop=list(data['Crop'].unique())):
    crop_data = data[data['Crop'] == crop]
    min_values = crop_data.min(numeric_only=True)
    max_values = crop_data.max(numeric_only=True)
    mean_values = crop_data.mean(numeric_only=True)
    
    print(f"Summary for {crop}:")
    print(f"Minimum values:\n{min_values}")
    print(f"Maximum values:\n{max_values}")
    print(f"Mean values:\n{mean_values}")
```

```python
@interact
def compare(condition=['Nitrogen', 'Phosphorus', 'Potassium', 'Temperature', 'Humidity', 'pH_Value', 'Rainfall']):
    overall_mean = data[condition].mean()
    print(f'Overall average value for {condition} is {overall_mean:.2f}')
    print('---------------------------------------------')
    for crop in data['Crop'].unique():
        crop_mean = data[data['Crop'] == crop][condition].mean()
        print(f'{crop}: {crop_mean:.2f}')
```

```python
@interact
def compare(condition=['Nitrogen', 'Phosphorus', 'Potassium', 'Temperature', 'Humidity', 'pH_Value', 'Rainfall']):
    print(f'Crops which require greater than average {condition}:')
    print(data[data[condition] >= data[condition].mean()]['Crop'].unique())
    print('---------------------------------------')
    print(f'Crops which require less than average {condition}:')
    print(data[data[condition] <= data[condition].mean()]['Crop'].unique())
```

```python
from sklearn.cluster import KMeans
```

```python
x = data.drop(columns=['Crop'])
x = x.values
x.shape
```

```python
inertia = []
K = range(1, 11)  # Testing k from 1 to 10
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)

# Plot the elbow chart
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```
